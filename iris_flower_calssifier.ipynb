```python
x = data.drop(['Species'], axis=1)
```

```python
x.head()
```

```python
y = data['Species']
```

```python
y.head()
```



**importing the model pre requirments :**



```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2 , random_state = 32)
```

```python
x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

**Logistic Regression model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_label = model.predict(x_test)
```

```python
y_label
```

```python
accuracy = accuracy_score(y_test,y_label)*100
print("the accuracy of the model is : ", accuracy , '%')
```

**SVM model**

```python
from sklearn import svm
```

```python
svm_model = svm.SVC()
svm_model.fit(x_train,y_train)
```

```python
y_predict = svm_model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
my_accuracy = accuracy_score(y_test,y_predict)*100
```

```python
print("the accuracy of svm model is : ", my_accuracy,'%')
```

**Decision tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_model = DecisionTreeClassifier()
dtc_model.fit(x_train,y_train)
```

```python
predection = dtc_model.predict(x_test)
```

```python
predection
```

```python
des_accuracy = accuracy_score(y_test,predection)*100
```

```python
print("the accuracy of decision tree model is : ", des_accuracy ,'%')
```

**Random forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(x_train,y_train)
```

```python
pred = forest.predict(x_test)
fore_accuracy = accuracy_score(y_test,pred)*100
print("accuracy is : ",fore_accuracy,'%')
```

```python
results = pd.DataFrame({
    'Model' : [
        'Logistic Regression' , 'SVM model' , 'Decision Tree' , 'Random Forest'
    ] ,
    'Accuracies' : [
        accuracy , my_accuracy, des_accuracy , fore_accuracy
    ]
})
```
